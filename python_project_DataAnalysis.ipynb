# Exploratory Data Analysis (EDA)

### 1- Comparing purchases of Male and Female

```python
df.columns
```

```python
gp = sns.countplot(x = 'Gender',data = df, order=df['Gender'].value_counts(ascending=True).index)

#If we write the above line only then also graph will show but it will be without label
#For calculating value of both

for bars in gp.containers:
    gp.bar_label(bars)
```

```python
# plotting a bar chart for gender vs total amoun
sales_amt = df.groupby(['Gender'],as_index = False)['Amount'].sum().sort_values(by = 'Amount' , ascending = True)

#If we write above line only then the data of male and female will show the total amount of money spend by them 
sns.barplot(x = 'Gender', y = 'Amount', data = sales_amt)
```

### From above two graph we can say that most number of buyers are Female and purchasing power of Female is higher than Male

### 2 - Comparing Different AGE Groups

```python
df.columns
```

```python
#Ploting graph for different age group and with bifergation of male and female
gp = sns.countplot(x = 'Age Group', hue = 'Gender',data = df, order=df['Age Group'].value_counts(ascending=True).index)
for bars in gp.containers:
    gp.bar_label(bars)
```

```python
# Total Amount spend by different Age Group
total_amt = df.groupby(['Age Group'],as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending = True)
sns.barplot(x = 'Age Group', y = 'Amount', data = total_amt)
```

### From above two Graph we can say the most number of buyers belongs to the 26-35yrs AGE Group and in that also Female has highest number

## 3 - Comparing sales of Different State

```python
df.columns
```

```python
#Ploting total numbers of orders placed by different state

sales_state = df.groupby(['State'], as_index = False)['Orders'].sum().sort_values(by = 'Orders', ascending = False)
#If we add .head() at the end of above line then it show graph of top 5 state

sns.set(rc = {'figure.figsize':(25,7)})
sns.barplot(data = sales_state, x = 'State', y = 'Orders')
```

```python
sales_state = df.groupby(['State'], as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending = False)

sns.set(rc = {'figure.figsize':(25,7)})
sns.barplot(data = sales_state, x = 'State', y = 'Amount')
```

### From above graphs we can see that most of the orders & total sales/amount are from Uttar Pradesh, Maharashtra and Karnataka respectively

### 4 - Comparing Marital Status 

```python
df.columns
```

```python
gp = sns.countplot(data = df, x = 'Marital_Status')

sns.set(rc = {'figure.figsize' : (2,3)})
for bars in gp.containers:
    gp.bar_label(bars)
```

```python
sales_amt = df.groupby(['Marital_Status', 'Gender'], as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending = False)

sns.set(rc = {'figure.figsize' : (6,5)})
sns.barplot(data = sales_amt, x = 'Marital_Status', y = 'Amount', hue = 'Gender')
```

### From above graphs we can see that most of the buyers are married (women) and they have high purchasing power

### 5 - Comparing Different Occupation

```python
df.columns
```

```python
sns.set(rc = {'figure.figsize':(20,5)})

gp = sns.countplot(data = df, x = 'Occupation')

for bars in gp.containers:
    gp.bar_label(bars)
```

```python
sales_amt = df.groupby(['Occupation'], as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending = True)

sns.set(rc = {'figure.figsize':(20,5)})
sns.barplot(data = sales_amt, x = 'Occupation', y = 'Amount')
```

### From above graphs we can see that most of the buyers are working in IT, Healthcare and Aviation sector

### 6 - Comparing Product Category

```python
df.columns
```

```python
sns.set(rc = {'figure.figsize':(25,10)})
gp = sns.countplot(data = df, x = 'Product_Category')

for bars in gp.containers:
    gp.bar_label(bars)
```

```python
sales_amt = df.groupby(['Product_Category'], as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending=True)

sns.set(rc = {'figure.figsize':(25,12)})
sns.barplot(data = sales_amt, x = 'Product_Category',y = 'Amount')
```

### From above graphs we can see that most of the sold products are from Food, Clothing and Electronics category
